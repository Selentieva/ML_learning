1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 10 * x - 5

def f_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

x_vals = np.linspace(0, 10, 100)
f_vals = f(x_vals)
f_derivative_vals = f_derivative(f, x_vals)

plt.plot(x_vals, f_vals, label='f(x) = 10x - 5', color='blue')
plt.plot(x_vals, f_derivative_vals, label="f'(x) = 10", color='red')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Залежність оцінки від кількості годин навчання є лінійною: з кожною додатковою годиною навчання оцінка зростає на 10 балів.


2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp
```

```python
x = sp.Symbol('x')

f_sin = sp.sin(x)
df_sin = sp.diff(f_sin, x)
print("Функція:", f_sin)
print("Похідна:", df_sin)

f_ln = sp.ln(x)
df_ln = sp.diff(f_ln, x)
print("Функція:", f_ln)
print("Похідна:", df_ln)

f_sin_func = sp.lambdify(x, f_sin, 'numpy')
df_sin_func = sp.lambdify(x, df_sin, 'numpy')
f_ln_func = sp.lambdify(x, f_ln, 'numpy')
df_ln_func = sp.lambdify(x, df_ln, 'numpy')

x_vals_sin = np.linspace(-2 * np.pi, 2 * np.pi, 200)
x_vals_ln = np.linspace(0.1, 10, 200)  

f_sin_vals = f_sin_func(x_vals_sin)
df_sin_vals = df_sin_func(x_vals_sin)
f_ln_vals = f_ln_func(x_vals_ln)
df_ln_vals = df_ln_func(x_vals_ln)

plt.plot(x_vals_sin, f_sin_vals, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals_sin, df_sin_vals, label="f'(x) = cos(x)", color='red', linestyle='--')
plt.plot(x_vals_ln, f_ln_vals, label='f(x) = ln(x)', color='brown')
plt.plot(x_vals_ln, df_ln_vals, label="f'(x) = 1/x", color='green', linestyle='--')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.Symbol('x')

f = x**2 - 5*x + 1
df = sp.diff(f, x)

f_func = sp.lambdify(x, f, 'numpy')
df_func = sp.lambdify(x, df, 'numpy')

x_vals = np.linspace(0, 100, 200)

f_vals = f_func(x_vals)
df_vals = df_func(x_vals)

plt.plot(x_vals, f_vals, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x_vals, df_vals, label="f'(x) = 2x - 5", color='red')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
